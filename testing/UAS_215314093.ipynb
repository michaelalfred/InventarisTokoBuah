```python
df_processed = df.drop('Countries and areas', axis=1)
df_processed.info()
```

```python
# Mengatur subplot dengan dua baris dan satu kolom
fig, axes = plt.subplots(nrows=len(df_processed.columns), ncols=1, figsize=(20, 25))

# Menampilkan histogram untuk setiap kolom
for i, kolom in enumerate(df_processed.columns):
    axes[i].hist(df_processed[kolom], bins=10)
    axes[i].set_title(kolom)
    axes[i].set_xlabel('Nilai')
    axes[i].set_ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```
